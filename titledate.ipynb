```python
title_list5 = []
date_list5 = []
driver2 = selenium.webdriver.Chrome()

for i in range(len(df5)):
    url = df5['url'].loc[i]
    driver2.get(url)
    source = driver2.page_source
    soup = BeautifulSoup(source)
    title = soup.find("h1")
    title = title.string
    title = str(title.string)
    title_list5.append(title)
    sourcedate = soup.find("div", class_="d2txt_1 clearfix")
    sourcedate = sourcedate.string
    sourcedate = str(sourcedate.string)
    date_list5.append(sourcedate)
    
print(len(title_list5))
print(len(date_list5))
```

```python
df5['title'] = title_list5
df5['source-date'] = date_list5
```

```python
import googletrans
from googletrans import Translator 
```

```python
ts = Translator()
```

```python
title_trans5 = []
for i in range(len(title_list5)):
    eng = ts.translate(title_list5[i])
    engtxt = eng.text
    title_trans5.append(engtxt)
len(title_trans5)
```
